```python
#  coffee machine needs to report what coffee is being delivered
#  how much resources are left, ie milk, coffee, water
#  resources need to adjust based on how much is consumed per order
#  keeps track of how much money has been placed into the machine
#  if pay for order is less than price, money is refunded to user
```

```python
global MENU, coins, resources
MENU = {
    "espresso": {
        "ingredients": {
            "water": 50,
            "coffee": 18,
        },
        "cost": 1.5,
    },
    "latte": {
        "ingredients": {
            "water": 200,
            "milk": 150,
            "coffee": 24,
        },
        "cost": 2.5,
    },
    "cappuccino": {
        "ingredients": {
            "water": 250,
            "milk": 100,
            "coffee": 24,
        },
        "cost": 3.0,
    }
}
coins = {
    "quarters":0.25,
    "dimes":0.10,
    "nickels":0.05,
    "pennies":0.01
}
resources = {
    "water": 300,
    "milk": 200,
    "coffee": 100,
}
```

```python
def coin_tracker(quarter_amount, dime_amount, nickel_amount, penny_amount):
    net_total = [(quarter_amount*coins["quarters"]), (dime_amount*coins["dimes"]), (nickel_amount*coins["nickels"]), (penny_amount*coins["pennies"])]

    return round(sum(net_total), 2)
```

```python
def coffee_maker(user_input):
    
    if user_input == "espresso":
        water_used = MENU[user_input]["ingredients"]["water"]
        coffee_used = MENU[user_input]["ingredients"]["coffee"]
        return water_used, coffee_used
    elif user_input == "latte":
        water_used = MENU[user_input]["ingredients"]["water"]
        coffee_used = MENU[user_input]["ingredients"]["coffee"]
        milk_used = MENU[user_input]["ingredients"]["milk"]
        return water_used, coffee_used, milk_used
    else:
        water_used = MENU[user_input]["ingredients"]["water"]
        coffee_used = MENU[user_input]["ingredients"]["coffee"]
        milk_used = MENU[user_input]["ingredients"]["milk"]
        return water_used, coffee_used, milk_used
    
```

```python
coffee_options = ["espresso", "latte", "cappuccino"]
report_options = ["off", "report"]
net_water = resources["water"]
net_milk = resources["milk"]
net_coffee = resources["coffee"]
stored_money = []

while True:
    user_input = input("What would you like? (espresso/latte/cappuccino): ")

    if user_input in coffee_options:
        coffee_cost = MENU[user_input]["cost"]
        user_pay = coin_tracker(int(input("How many quarters?")), int(input("How many dimes?")), int(input("How many nickels?")), int(input("How many pennies?")))
        if user_input == "espresso":
            if net_water < MENU[user_input]["ingredients"]["water"]:
                print("Not enough water left.")
                continue
            elif net_coffee < MENU[user_input]["ingredients"]["coffee"]:
                print("Not enough coffee left.")
                continue
            else:
                if coffee_cost > user_pay:
                    print("Sorry, that's not enough money. Money refunded")
                else:
                    change = user_pay - coffee_cost
                    used_water, used_coffee = coffee_maker(user_input)
                    net_water = net_water - used_water
                    net_coffee = net_coffee - used_coffee
                    stored_money.append(MENU[user_input]["cost"])
                    print(f"Here is your ${round(change, 2)} in change.\nHere is your {user_input} ☕ Enjoy!")
        else:
            if user_input == "latte":
                if net_water < MENU[user_input]["ingredients"]["water"]:
                    print("Not enough water left.")
                    continue
                elif net_milk < MENU[user_input]["ingredients"]["milk"]:
                    print("Not enough milk left.")
                    continue
                elif net_coffee < MENU[user_input]["ingredients"]["coffee"]:
                    print("Not enough coffee left.")
                    continue
                else:
                    if coffee_cost > user_pay:
                        print("Sorry, that's not enough money. Money refunded.")
                    else:
                        change = user_pay - coffee_cost
                        used_water, used_coffee, used_milk = coffee_maker(user_input)
                        net_water = net_water - used_water
                        net_coffee = net_coffee - used_coffee
                        stored_money.append(MENU[user_input]["cost"])
                        print(f"Here is your ${round(change, 2)} in change.\nHere is your {user_input} ☕ Enjoy!")
            if user_input == "cappuccino":
                if net_water < MENU[user_input]["ingredients"]["water"]:
                    print("Not enough water left.")
                    continue
                elif net_milk < MENU[user_input]["ingredients"]["milk"]:
                    print("Not enough milk left.")
                    continue
                elif net_coffee < MENU[user_input]["ingredients"]["coffee"]:
                    print("Not enough milk left.")
                    continue
                else:
                    if coffee_cost > user_pay:
                        print("Sorry, that's not enough money. Money refunded")
                    else:
                        change = user_pay - coffee_cost
                        used_water, used_coffee, used_milk = coffee_maker(user_input)
                        net_water = net_water - used_water
                        net_coffee = net_coffee - used_coffee
                        stored_money.append(MENU[user_input]["cost"])
                        print(f"Here is your ${round(change, 2)} in change.\nHere is your {user_input} ☕ Enjoy!")
    elif user_input == "report":
        print(f"Water: {net_water}mL\nMilk: {net_milk}mL\nCoffee: {net_coffee}g\nMoney: ${round(sum(stored_money), 2)}")
        continue
    elif user_input == "off":
        print("Turning off...")
        break
    else:
        print("Invalid  input.")
        continue
```
